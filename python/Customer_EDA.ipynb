# 2. Data Cleaning & Preprocessing

## 2.1 Drop Redundant Columns

```python
# Drop the customer_id column
df = df.drop('customer_id', axis=1)
df.info()
```

```python
# Rename purchase_amount_usd column
df.rename(columns={'purchase_amount_usd': 'purchase_amount'}, inplace=True)
```

## 2.2 Process Columns from Text to Binary

```python
# Map Yes/No to 1/0 for binary columns
binary_cols = ['subscription_status', 'discount_applied', 'promo_code_used']
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

df.head()
```

## 2.3 Data Check

```python
# Final validation
print("Cleaned dataset info:")
df.info()
print(f"\nFinal shape: {df.shape}")
print(f"Missing values:\n{df.isnull().sum()}")
```

# 3. Explaratory Data Analysis

```python
# Separate different variables for analysis
categorical_columns = ['gender', 'item_purchased', 'category', 'location',
                       'size', 'color','shipping_type', 'payment_method', 'frequency_of_purchases', 'season']

numerical_columns = ['age', 'purchase_amount', 'review_rating', 'subscription_status',
                     'discount_applied', 'promo_code_used', 'previous_purchases']
```

## 3.1 Univariate Analysis

### 3.1.1 Categorical Columns

```python
# Create subplots for visualizing all categorical columns
plt.figure(figsize=(20, 25))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(5, 2, i)
    ax = sns.countplot(data=df, x=col, hue=col, order=df[col].value_counts().index, palette="Set3", legend=False)
    plt.xticks(rotation=45)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

### 3.1.2 Numerical Columns

```python
# Columns with continuous numerical values
con_num_col = ['age', 'purchase_amount', 'review_rating', 'previous_purchases']
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

for i, col in enumerate(con_num_col):
    # Histogram
    plt.subplot(len(con_num_col), 2, 2*i+1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Boxplot
    plt.subplot(len(con_num_col), 2, 2*i+2)
    sns.boxplot(data=df, x=col, color='lightgreen')
    plt.title(f'{col} Boxplot')

plt.tight_layout()
plt.show()
```

## 3.2 Bivariate Analysis

### 3.2.1 Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix (Numerical Features)")
plt.tight_layout()
plt.show()
```

### 3.2.2 Group-Based Insights

```python
# Average Purchase by Category
category_avg = df.groupby("category")["purchase_amount"].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=category_avg.index, hue=category_avg.index, y=category_avg.values, palette="Set3", legend=False)
plt.title("Average Purchase Amount by Category")
plt.xlabel("Category")
plt.ylabel("Average Purchase Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Purchases by Gender
gender_counts = df["gender"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_counts.index, y=gender_counts.values, hue=gender_counts.index, palette="pastel", legend=False)
plt.title("Total Purchases by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Purchases")
plt.show()
```

```python
# Average Review Rating by Payment Method
payment_avg_review = df.groupby("payment_method")["review_rating"].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=payment_avg_review.index, y=payment_avg_review.values, hue=payment_avg_review.index, palette="coolwarm", legend=False)
plt.title("Average Review Rating by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Previous Purchases by Frequency
freq_loyalty = df.groupby("frequency_of_purchases")["previous_purchases"].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=freq_loyalty.index, y=freq_loyalty.values, hue=freq_loyalty.index, palette="Blues", legend=False)
plt.title("Avg Previous Purchases by Frequency of Purchases")
plt.xlabel("Frequency")
plt.ylabel("Avg Previous Purchases")
plt.xticks(rotation=45)
plt.show()
```
